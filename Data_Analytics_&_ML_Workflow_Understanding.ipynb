```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating healthcare dataset directly (no CSV required)

data = {
    "Age": [45, 34, 50, 29, 60, 41, 36, 55],
    "BMI": [28.5, 22.1, 31.2, 21.5, 33.8, 26.4, 24.8, 30.1],
    "BloodPressure": [130, 120, 140, 110, 150, 125, 118, 145],
    "Glucose": [160, 98, 170, 90, 190, 130, 105, 175],
    "Cholesterol": [220, 180, 250, 170, 270, 210, 190, 260],
    "PhysicalActivity": ["Low", "High", "Low", "High", "Low", "Medium", "Medium", "Low"],
    "Smoking": ["Yes", "No", "Yes", "No", "Yes", "No", "No", "Yes"]
}

df = pd.DataFrame(data)

print("Dataset loaded successfully!")
df.head()
```

```python
print("Dataset Information:")
df.info()

print("\nStatistical Summary:")
df.describe()
```

```python
print("Missing Values:")
df.isnull().sum()
```

```python
print("Business Context:")
print("Dataset represents patient health indicators.")
print("Objective: Analyze data and understand how ML problems are framed.")
```

```python
plt.figure()
plt.hist(df["Age"], bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.scatter(df["BMI"], df["Glucose"])
plt.title("BMI vs Glucose Level")
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="Smoking", y="Cholesterol", data=df)
plt.title("Cholesterol Levels by Smoking Status")
plt.show()
```

```python
plt.figure()
numeric_data = df.select_dtypes(include=np.number)
sns.heatmap(numeric_data.corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
print("Feature Identification:")
print("Input Features:")
for col in df.columns:
    print("-", col)

print("\nPotential Target Variable (Future Scope):")
print("- Health Risk Level (Low / Medium / High)")
```
